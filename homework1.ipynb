```python
import os
import nltk
import re
from nltk.corpus import stopwords
nltk.download('punkt')
```

Загружаем файлы с отзывами с imdb(взял отсюда: https://ai.stanford.edu/~amaas/data/sentiment/)

```python
good_rev_files = os.listdir('movie_pos')
bad_rev_files = os.listdir('movie_neg')
good_rev =[]
bad_rev = []
```

```python
for filename in good_rev_files:
    try:
        f = open('movie_pos/'+filename, "r",)
        good_rev.append(f.read())
    except UnicodeDecodeError:
        continue
```

```python
for filename in bad_rev_files:
    try:
        f = open('movie_neg/'+filename, "r")
        bad_rev.append(f.read())
    except UnicodeDecodeError:
        continue
```

Создаю множества и словари с словами из хороших и плохих отзывов

```python
good_un_words = set()
good_word_dict = {}
for review in good_rev:
    review_words = nltk.word_tokenize(review)
    review_words =[word.lower() for word in review_words if word.isalpha()]
    for word in review_words:
        good_un_words.add(word)
    for word in review_words:
        if word in good_word_dict.keys():
            good_word_dict[word] += 1
        else:
            good_word_dict[word] = 0
```

```python
bad_un_words = set()
bad_word_dict = {}
for review in bad_rev:
    review_words = nltk.word_tokenize(review)
    review_words =[word.lower() for word in review_words if word.isalpha()]
    for word in review_words:
        bad_un_words.add(word)
    for word in review_words:
        if word in bad_word_dict.keys():
            bad_word_dict[word] += 1
        else:
            bad_word_dict[word] = 0
```

Множества слов, которые встречаются только в хороших и только в плохих отзывах при этом встречаются 10 и больше раз

```python
a = good_un_words.difference(bad_un_words)
good_un = set()
for word in a:
    if good_word_dict[word]>10:
        good_un.add(word)
```

```python
b = bad_un_words.difference(good_un_words)
bad_un = set()
for word in b:
    if bad_word_dict[word]>10:
        bad_un.add(word)
```

Тестируем

```python
rev_files = os.listdir('test_data')
test_rev = []
for filename in rev_files:
    f = open('test_data/'+filename, "r")
    test_rev.append(f.read())
```

```python
for review in test_rev:
    pos=0
    neg=0
    review_words = nltk.word_tokenize(review)
    review_words =[word.lower() for word in review_words if word.isalpha()]
    for word in review_words:
        if word in a_un:
            pos+=1
        if word in b_un:
            neg+=1
    print(review+'\n')
    print('Positive words: '+str(pos)+'   '+'Negative words: '+str(neg))
    if pos >= neg:
        print('positive')
    else:
        print('negative')
    print('\n\n')
```
